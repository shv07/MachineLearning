```python
import random
from matplotlib import pyplot as plt
import numpy as np
#refernece: https://www.numpy.org/devdocs/reference/generated/numpy.linspace.html
#reference: https://www.scipy-lectures.org/intro/numpy/auto_examples/plot_polyfit.html
```

```python
def create_dataset(m=5,c=10):
    X=list(range(-100,100,4))
    ch=range(-300,300)
    Y=[]
    Y1=[]

    for i in range(50):
        y=m*X[i]+c
        Y.append(y)
        y_=y+0.7*random.choice(ch)
        Y1.append(y_)
    return [X,Y1,Y]
[X,Y1,Y]=create_dataset()
x=X
y1=Y1
y=Y
#plt.scatter(x,y1,'b')
plt.plot(x,y,'r')
plt.scatter(x,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Data Plot")
plt.show()

```

```python
x = X
y = Y1
p = np.poly1d(np.polyfit(x, y, 5))  #5 deg polynomial fitting of the data

t = np.linspace(x[0],x[-1],500)     #returns evenly spaced 500 numbers in the range x[0], x[-1]
plt.plot(x, y, 'o', t, p(t), '-')  #p(t) returns the values of polynomial function corresponding to values of t
plt.xlabel("X")
plt.ylabel("Y")
plt.title("5 degree poly fit")
plt.show()
```
