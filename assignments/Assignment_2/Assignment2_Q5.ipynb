## Q.5) a & b : Random Number Generator

```python
#refernce: https://en.wikipedia.org/wiki/Linear_congruential_generator 
import datetime
import random
from matplotlib import pyplot as plt
#linear congruential generator
#generates a random number in the range 1-n 
def r_gen(n):
    now=datetime.datetime.now()
    X=int(now.microsecond)%1000
    tmp=X
    m=2**32
    a=214013
    c=2531011
    r_list=[X]
    for i in range(n):
        X=(a*X+c)%m
        r_list.append(X)
    return ((r_list[int(n-1)])%n+1)
R=[]
x=[]
for i in range(1000):
    R.append(r_gen(100))
    x.append(i)
y=R
plt.plot(x,y,'r')
plt.xlabel("Iteration No.")
plt.ylabel("Random Number generated")
plt.title("Random Number by LCG algorithm")
plt.show()
```

```python
plt.hist(y)
```

## Equivalent program with the inbuilt random.choice() function

```python
r_no1=[]
X1=[]
for i in range(1000):
    x=random.choice(range(1,100))
    r_no1.append(x)
    X1.append(i)
x=X1
y=r_no1
plt.plot(x,y,'r')
plt.xlabel("Iteration No.")
plt.ylabel("Random Number generated")
plt.title("Random Number by inbuilt Function")
plt.show()
plt.hist(y)
```
