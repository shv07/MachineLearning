## Q.5) a & b : Random Number Generator

```python
import datetime
import random
from matplotlib import pyplot as plt
r_no=[]
X=[]
for i in range(1000):
    now=datetime.datetime.now()
    x=int(now.microsecond)
    y=int(now.microsecond)
    x=int((x+y)%(100))
    r_no.append(x)
    X.append(i)
x=X
y=r_no
plt.plot(x,y,'r')
plt.xlabel("Unique Bigram Idex")
plt.ylabel("Count before add-1 smoothing")
plt.title("Count v/s bigram index")
plt.show()
plt.hist(y)
```

```python
r_no1=[]
X1=[]
for i in range(1000):
    x=random.choice(range(1,100))
    r_no1.append(x)
    X1.append(i)
x=X1
y=r_no1
plt.plot(x,y,'r')
plt.xlabel("Unique Bigram Idex")
plt.ylabel("Count before add-1 smoothing")
plt.title("Count v/s bigram index")
plt.show()
plt.hist(y)
```

```python
#linear congruential generator
#generates a random number in the range 1-n 
def r_gen(n):
    now=datetime.datetime.now()
    X=int(now.microsecond)%1000
    tmp=X
    m=2**32
    a=214013
    c=2531011
    r_list=[X]
    for i in range(n):
        X=(a*X+c)%m
        r_list.append(X)
    return ((r_list[int(n-1)])%n+1)
R=[]
x=[]
for i in range(1000):
    R.append(r_gen(100))
    x.append(i)
y=R
plt.plot(x,y,'r')
plt.xlabel("Iteration No.")
plt.ylabel("Count before add-1 smoothing")
plt.title("Count v/s bigram index")
plt.show()
```
